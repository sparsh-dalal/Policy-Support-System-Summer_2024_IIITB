```python
import pandas as pd
import numpy as np
```

```python
import re
def remove_pattern(col_name):
    return re.sub(r'_[0-9]{3}$', '', col_name)

# I am doing this because coloumn names are to be cleaned before accessing.
```

```python
Zones = pd.read_excel('Zones.xlsx')
Dataset = pd.read_csv('D:\IIIT-B\WSL_PSS\KAG 2016-17\Agriculture\Agriculture_KAG_2016_17.csv')
Dataset.columns = map(remove_pattern, Dataset.columns)
Working_Directory = Dataset[['District','TotalNPK', 'Rice_Production','Maize_Production']].copy()
Working_Directory['Admin_Zone'] = Zones['Administrative Region']

Working_Directory
```

```python
Working_Directory['TotalNPK'][24] = int(Working_Directory['TotalNPK'][24]/2)
Working_Directory['Rice_Production'][24] = int(Working_Directory['Rice_Production'][24]/2)
Working_Directory['Maize_Production'][24] = Working_Directory['Maize_Production'][24]/2
Working_Directory

# This is done to divide values of Ballari to 2, for new district Vijaynagra for which data is not available
```

```python
Working_Directory.loc[30] = {
    'District': 'Vijayanagara',
    'Admin_Zone': Working_Directory['Admin_Zone'][24],
    'TotalNPK': int(Working_Directory['TotalNPK'][24]),
    'Rice_Production': int(Working_Directory['Rice_Production'][24]),
    'Maize_Production': Working_Directory['Maize_Production'][24]
}

Working_Directory
```

```python
R_NPK_1 = []
R_NPK_2 = []
R_NPK_3 = []
R_NPK_4 = []
M_NPK_1 = []
M_NPK_2 = []
M_NPK_3 = []
M_NPK_4 = []
Rice_1 = []
Rice_2 = []
Rice_3 = []
Rice_4 = []
Maize_1 = []
Maize_2 = []
Maize_3 = []
Maize_4 = []
for i in range(len(Working_Directory['Admin_Zone'])):
    if Working_Directory['Admin_Zone'][i] == 'Bengaluru Division':
        Total = Working_Directory['Rice_Production'][i] + Working_Directory['Maize_Production'][i]
        
        
        NPK_1.append(Working_Directory['TotalNPK'][i])
        Rice_1.append(Working_Directory['Rice_Production'][i])
        Maize_1.append(Working_Directory['Maize_Production'][i])
    elif Working_Directory['Admin_Zone'][i] == 'Mysuru Division':
        NPK_2.append(Working_Directory['TotalNPK'][i])
        Rice_2.append(Working_Directory['Rice_Production'][i])
        Maize_2.append(Working_Directory['Maize_Production'][i])
    elif Working_Directory['Admin_Zone'][i] == 'Belagavi Division':
        NPK_3.append(Working_Directory['TotalNPK'][i])
        Rice_3.append(Working_Directory['Rice_Production'][i])
        Maize_3.append(Working_Directory['Maize_Production'][i])
    else:
        NPK_4.append(Working_Directory['TotalNPK'][i])
        Rice_4.append(Working_Directory['Rice_Production'][i])
        Maize_4.append(Working_Directory['Maize_Production'][i])


# Doing this to Divide my X and Y data points for 4 models. Can be skipped, but a segregated data might be useful in the future?
```

```python
import scipy

m1_rice = scipy.stats.linregress(NPK_1, Rice_1)
m2_rice = scipy.stats.linregress(NPK_2, Rice_2)
m3_rice = scipy.stats.linregress(NPK_3, Rice_3)
m4_rice = scipy.stats.linregress(NPK_4, Rice_4)
m1_maize = scipy.stats.linregress(NPK_1, Maize_1)
m2_maize = scipy.stats.linregress(NPK_2, Maize_2)
m3_maize = scipy.stats.linregress(NPK_3, Maize_3)
m4_maize = scipy.stats.linregress(NPK_4, Maize_4)

M_Rice = []
M_Maize = []

for i in range(len(Working_Directory['Admin_Zone'])):
    if Working_Directory['Admin_Zone'][i] == 'Bengaluru Division':
        M_Rice.append(m1_rice.slope)
        M_Maize.append(m1_maize.slope)
    elif Working_Directory['Admin_Zone'][i] == 'Mysuru Division':
        M_Rice.append(m2_rice.slope)
        M_Maize.append(m2_maize.slope)
    elif Working_Directory['Admin_Zone'][i] == 'Belagavi Division':
        M_Rice.append(m3_rice.slope)
        M_Maize.append(m3_maize.slope)
    else:
        M_Rice.append(m4_rice.slope)
        M_Maize.append(m4_maize.slope)


Working_Directory['M_Rice'] = M_Rice
Working_Directory['M_Maize'] = M_Maize

Working_Directory

# Linear Regression Model. Storing the value of slope in the dataframe itself.
```

```python
def calculation_new_values_Rice(pertubation, Working_Directory, coloumn_name, m_value):
    New_values = []
    for i in range(len(Working_Directory['District'])):
        y_new = coloumn_name[i] + m_value[i] * (pertubation * Working_Directory['TotalNPK'][i]/100)
        New_values.append(int(y_new))
    Working_Directory[f'R - NPK {pertubation}'] = New_values

def calculation_new_values_Maize(pertubation, Working_Directory, coloumn_name, m_value):
    New_values = []
    for i in range(len(Working_Directory['District'])):
        y_new = coloumn_name[i] + m_value[i] * (pertubation * Working_Directory['TotalNPK'][i]/100)
        New_values.append(int(y_new))
    Working_Directory[f'M - NPK {pertubation}'] = New_values


# 2 Functions, to calculate new values for change in NPK Production.
```
