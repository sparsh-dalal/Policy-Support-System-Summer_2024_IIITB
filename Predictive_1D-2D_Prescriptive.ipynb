```python
import pandas as pd
import numpy as np

adjacency_excel = pd.read_excel('D:\IIIT-B\WSL_PSS\Karnataka_District_Adjacency_File - Copy.xlsx')

district_codes = sorted(set(adjacency_excel['District_Code']))

n = len(district_codes)
adj_matrix = np.zeros((n, n), dtype=int)

code_to_index = {code: idx for idx, code in enumerate(district_codes)}

for i in range(len(adjacency_excel)):
    district_code = adjacency_excel['District_Code'][i]
    neighbors = [int(num) for num in str(adjacency_excel['District_Neighbors'][i]).split(',') if num.strip().isdigit()]
    
    for neighbor in neighbors:
        adj_matrix[code_to_index[district_code]][code_to_index[neighbor]] = 1
        adj_matrix[code_to_index[neighbor]][code_to_index[district_code]] = 1  # Assuming adjacency is bidirectional

adj_matrix_df = pd.DataFrame(adj_matrix, index=district_codes, columns=district_codes)

print(adj_matrix_df)

adj_matrix_np = adj_matrix_df.to_numpy()
```

```python
def normalize_list(lst):
    min_val = min(lst)
    max_val = max(lst)
    normalized_lst = [(x - min_val) / (max_val - min_val) for x in lst]
    return normalized_lst
```

```python
import numpy as np

def calculate_avg_capability(capability, neighbors):
    num_neighbors = len(neighbors)
    if num_neighbors == 0:
        return 0
    sum_capabilities = sum(capability[n] for n in neighbors)
    return sum_capabilities / num_neighbors

def calculate_stress_stability(capability, adjacency_matrix):
    n = len(capability)
    stress = np.zeros(n)
    stability = np.zeros(n)
    
    for i in range(n):
        neighbors = [j for j in range(n) if adjacency_matrix[i][j] == 1 and i != j]
        if len(neighbors) > 0:
            avg_capability = calculate_avg_capability(capability, neighbors)
            stress[i] = abs(avg_capability - capability[i])
            stability[i] = 1 - stress[i]
        else:
            stress[i] = 0  # If no neighbors, stress is zero
            stability[i] = 1
        
    return stress, stability




stress, stability = calculate_stress_stability(Capability_list, adj_matrix_np)

print("Stress values:", stress)
print("Stability values:", stability)
```

```python
import numpy as np

# Capabiliy - 1/eigenvector(Capabilitty_other nodes)

def bonacich_bargain_centrality(districts, capabilities, adjacency_matrix):
    adjacency_matrix = np.array(adjacency_matrix)
    
    eigenvalues, _ = np.linalg.eig(adjacency_matrix)
    largest_eigenvalue = max(eigenvalues)
    
    beta = 1 / largest_eigenvalue
    
    b = np.array(capabilities, dtype=float)
    
    n = len(districts)
    
    
    for rec in range(0,10):
        b_old = b.copy()
            
        for i in range(n):
            sum_term = sum(adjacency_matrix[i, j] * b_old[j] for j in range(n) if i != j)
            b[i] = b_old[i] - beta * sum_term
            
        
        
    return b.tolist()

result = bonacich_bargain_centrality(Other_Directory['District'], Capability_list, adj_matrix_df)


result = normalize_list(result)
new_list = []
new_list = [1 - value for value in result] # stability
len(new_list)
```

```python
import numpy as np
import pandas as pd

def calculate_stress_stability(adj_matrix, df):
    # Initialize stress and stability columns
    df['stress'] = 0.0
    df['stability'] = 0.0
    
    # Extract the capability columns
    capability_columns = df.columns.difference(['node', 'stress', 'stability'])
    
    # Iterate through each node in the DataFrame
    for idx, row in df.iterrows():
        node = row['node']
        node_idx = df.index[df['node'] == node][0]
        
        # Find neighbors (where adj_matrix[node_idx] is 1)
        neighbors_idx = np.where(adj_matrix[node_idx] == 1)[0]
        
        if len(neighbors_idx) > 0:
            # Get the capability vectors of the neighbors
            neighbor_capabilities = df.iloc[neighbors_idx][capability_columns].values
            
            # Calculate the centroid of the neighbors' capability vectors
            centroid = np.mean(neighbor_capabilities, axis=0)
            
            # Calculate the L2 distance (stress) from the node's capability vector to the centroid
            node_capabilities = row[capability_columns].values
            stress = np.linalg.norm(node_capabilities - centroid)
            
            # Calculate stability
            stability = 1 - stress
        else:
            # If no neighbors, set stress to 0 and stability to 1 (arbitrary choice)
            stress = 0
            stability = 1
        
        # Update the DataFrame with the calculated values
        df.at[idx, 'stress'] = stress
        df.at[idx, 'stability'] = stability
    
    return df


D2_Capabilities = pd.DataFrame()

capability_1 = normalize_list(Other_Directory['R - TotalNPK 10'])
capability_2 = normalize_list(Other_Directory['M - TotalNPK 10'])

D2_Capabilities = {'node': Other_Directory['District'], 'Capability_1': capability_1, 'Capability_2': capability_2}
D2_Capabilities = pd.DataFrame(D2_Capabilities)


result_df = calculate_stress_stability(adj_matrix_np, D2_Capabilities)
print(result_df)
```

```python
def normalize_list(lst):
    min_val = min(lst)
    max_val = max(lst)
    normalized_lst = [(x - min_val) / (max_val - min_val) for x in lst]
    return normalized_lst
```

```python
import pandas as pd
import numpy as np

adjacency_excel = pd.read_excel('D:\IIIT-B\WSL_PSS\Karnataka_District_Adjacency_File - Copy.xlsx')

district_codes = sorted(set(adjacency_excel['District_Code']))

n = len(district_codes)
adj_matrix = np.zeros((n, n), dtype=int)

code_to_index = {code: idx for idx, code in enumerate(district_codes)}

for i in range(len(adjacency_excel)):
    district_code = adjacency_excel['District_Code'][i]
    neighbors = [int(num) for num in str(adjacency_excel['District_Neighbors'][i]).split(',') if num.strip().isdigit()]
    
    for neighbor in neighbors:
        adj_matrix[code_to_index[district_code]][code_to_index[neighbor]] = 1
        adj_matrix[code_to_index[neighbor]][code_to_index[district_code]] = 1  # Assuming adjacency is bidirectional

adj_matrix_df = pd.DataFrame(adj_matrix, index=district_codes, columns=district_codes)

print(adj_matrix_df)

adj_matrix_np = adj_matrix_df.to_numpy()
```

```python
import numpy as np

def calculate_avg_capability(capability, neighbors):
    num_neighbors = len(neighbors)
    if num_neighbors == 0:
        return 0
    sum_capabilities = sum(capability[n] for n in neighbors)
    return sum_capabilities / num_neighbors

def calculate_stress_stability(capability, adjacency_matrix):
    n = len(capability)
    stress = np.zeros(n)
    stability = np.zeros(n)
    
    for i in range(n):
        neighbors = [j for j in range(n) if adjacency_matrix[i][j] == 1 and i != j]
        if len(neighbors) > 0:
            avg_capability = calculate_avg_capability(capability, neighbors)
            stress[i] = abs(avg_capability - capability[i])
            stability[i] = 1 - stress[i]
        else:
            stress[i] = 0  # If no neighbors, stress is zero
            stability[i] = 1
        
    return stress, stability




# stress, stability = calculate_stress_stability(Capability_list, adj_matrix_np)

# print("Stress values:", stress)
# print("Stability values:", stability)
```
